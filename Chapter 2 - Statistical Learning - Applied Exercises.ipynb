Interesting, the horsepower variable's dtype is an object, even though it is clearly a numeric variable. It turns out that it contains '?', which turns it into an object. As it is clearly a mistake in the data (or lack of information for the vehicle), I will set every the value of every question mark to zero.

```python
auto['horsepower'] = np.where(auto['horsepower'] == '?', 0, auto['horsepower'])
```

```python
auto['horsepower'] = auto['horsepower'].astype('float64')
```

2) What is the range of each quantitative predictor?

```python
def variable_range(x: ndarray) -> Dict[str, ndarray]:
    
    var_range: Dict[str, ndarray] = {}
    var_range['min'] = np.min(x)
    var_range['max'] = np.max(x)
    
    return var_range
```

```python
for i in range(len(auto.columns)):
    if auto.iloc[:, i].dtype == 'float64' or auto.iloc[:, i].dtype == 'int64':
        print(auto.columns[i], ':', variable_range(auto.iloc[:, i])['min'], '-',
             variable_range(auto.iloc[:, i])['max'])
```

3) What is the mean and standard deviation of each quantitative predictor?

```python
for i in range(len(auto.columns)):
    if auto.iloc[:, i].dtype == 'float64' or auto.iloc[:, i].dtype == 'int64':
        print(f"{auto.columns[i]}, mean: {np.mean(auto.iloc[:, i])}, stdev: {np.std(auto.iloc[:, i])}")
```
